```python
# histogram for continuous variable "Age"
plt.figure(figsize=(10,5))
plt.hist(df['Age'],bins=10,color='b',edgecolor='k',alpha=0.5)
plt.title('Histogram for Patient Age')
plt.xlabel("Patient Age")
plt.ylabel('Frequency')
plt.show()
```

```python
#histogram for the continuous variable Na_to_K
plt.figure(figsize=(10,5))
plt.hist(df["Na_to_K"],bins=10,color='red',edgecolor='k',alpha=0.7)
plt.title("Histogram for Sodium to Pottasium ratio in Blood")
plt.xlabel("Na_to_k")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart for categorical variable BP
plt.figure(figsize=(10,5))
df['BP'].value_counts().plot(kind='bar',color=['red','yellow','green'],edgecolor='k',alpha=0.5)
plt.title("Barchart for the Blood Pressure of the Patients")
plt.xlabel("Blood pressure")
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
# Bar chart for categorical variable Sex
plt.figure(figsize=(10,5))
df['Sex'].value_counts().plot(kind='bar',color=['blue','red'],edgecolor='k',alpha=0.7)
plt.title("Barchart for the Gender of the Patients")
plt.xlabel("Gender")
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
# Bar chart for categorical variable Cholestrol
plt.figure(figsize=(10,5))
df['Cholesterol'].value_counts().plot(kind='bar',color=['purple','skyblue'],edgecolor='k',alpha=0.7)
plt.title("Barchart for the Cholesterol of the Patients")
plt.xlabel("Cholesterol")
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
# Bar chart for categorical variable Drug
plt.figure(figsize=(10,5))
df['Drug'].value_counts().plot(kind='bar', color=['red','green','yellow','orange','blue'],edgecolor='k',alpha=0.7)
plt.title("Barchart for the types of Drugs")
plt.xlabel("Drugs")
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
# Barchart for Gender and Drugs
# group the data
gender_drug_counts=df.groupby(['Drug','Sex']).size().unstack(fill_value=0)
#plot grouped bar chart
gender_drug_counts.plot(kind='bar',color=['blue','pink'], stacked=False,figsize=(10,5),edgecolor='k',alpha=0.7)
plt.xlabel('Drug Type')
plt.xticks(rotation=0)
plt.ylabel('Count')
plt.title('Barchart for Gender and Drug Type')
plt.show()
```

```python
# Barchart for Gender and BP
# group the data
gender_bp_counts=df.groupby(['BP','Sex']).size().unstack(fill_value=0)
#plot grouped bar chart
gender_bp_counts.plot(kind='bar',color=['r','g'], stacked=False,figsize=(10,5),edgecolor='k',alpha=0.7)
plt.xlabel('BP')
plt.ylabel('Count')
plt.title('Blood Pressure based on Gender')
plt.show()
```
